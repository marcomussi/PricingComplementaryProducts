```python
import numpy as np
import matplotlib.pyplot as plt

from efficientregressors import GaussianProcessRegressor
```

```python
kernel_L = 5
n_samples_plt = 200
x_plt = np.linspace(0, 1, n_samples_plt)

data_x = np.array([[0.2],[0.4],[0.8],[1.0]])
data_y = np.array([[0.2],[0.4],[0.8],[0.2]])
sigma_eq = np.array([[0.01],[0.2],[0.1],[0.5]])
hetRegressor = HeteroskedasticGaussianProcessRegressor(kernel_L, 1)
hetRegressor.load_data(data_x, data_y, sigma_eq)
y_hat, y_sigma = hetRegressor.compute(x_plt.reshape(-1, 1))
plt.plot(x_plt, y_hat, label="Heteroskedastic GP")
plt.fill_between(x_plt, y_hat - y_sigma, y_hat + y_sigma, alpha=0.3)
plt.scatter(data_x, data_y, label="Heteroskedastic GP Data")

data_x = np.array([[0.2],[0.4],[0.8],[0.8],[0.8],[1.0]])
data_y = np.array([[0.2],[0.4],[0.9],[0.9],[0.6],[0.2]])
sigma_eq = np.array([[0.01],[0.2],[0.3],[0.3],[0.3],[0.5]])
hetRegressor = HeteroskedasticGaussianProcessRegressor(kernel_L, 1)
hetRegressor.load_data(data_x, data_y, sigma_eq)
y_hat, y_sigma = hetRegressor.compute(x_plt.reshape(-1, 1))
plt.plot(x_plt, y_hat, label="Heteroskedastic GP 2")
plt.fill_between(x_plt, y_hat - y_sigma, y_hat + y_sigma, alpha=0.3)
plt.scatter(data_x, data_y, label="Heteroskedastic GP 2 Data")


plt.xlim([0, 1])
plt.legend()
```

```python
kernel_L = 5
base_sigma_sq = 0.25
n_samples_plt = 200
x_plt = np.linspace(0, 1, n_samples_plt)

plt.figure()

data_x = np.array([0.1, 0.3, 0.8]).reshape(-1, 1)
data_y = np.array([3/8, 3/5, 4/10]).reshape(-1, 1)
sigma_sq_eq = base_sigma_sq * np.array([1/16, 1/5, 1/10]).reshape(-1, 1)
hetRegressor = HeteroskedasticGaussianProcessRegressor(kernel_L, 1)
hetRegressor.load_data(data_x, data_y, sigma_sq_eq)
y_hat1, y_sigma1 = hetRegressor.compute(x_plt.reshape(-1, 1))
plt.plot(x_plt, y_hat1, color="blue", label="Heteroskedastic GP")
plt.fill_between(x_plt, y_hat1 - y_sigma1, y_hat1 + y_sigma1, color="blue", alpha=0.3)
plt.scatter(data_x, data_y, color="blue", label="Heteroskedastic GP Data")

data_x = np.vstack((0.1 * np.ones((16, 1)), 0.3 * np.ones((5, 1)), 0.8 * np.ones((10, 1))))
data_y = np.vstack((np.ones((6, 1)), np.zeros((10, 1)), np.ones((3, 1)), np.zeros((2, 1)), np.ones((4, 1)), np.zeros((6, 1))))
homRegressor = GaussianProcessRegressor(kernel_L, base_sigma_sq, 1)
homRegressor.load_data(data_x, data_y)
y_hat2, y_sigma2 = homRegressor.compute(x_plt.reshape(-1, 1))
plt.plot(x_plt, y_hat2, color="orange", label="Homoskedastic GP")
plt.fill_between(x_plt, y_hat2 - y_sigma2, y_hat2 + y_sigma2, color="orange", alpha=0.3)
plt.scatter(data_x, data_y, color="orange", label="Homoskedastic GP Data")

plt.xlim([0, 1])
plt.ylim([- 0.1, 1.1])
plt.legend()
```
